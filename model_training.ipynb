FEATURE ENGINEERING

```python
# ==========================================================
# STEP 1A: Create Student Master Table
# ==========================================================

student_master = (
    studentInfo_clean[
        [
            "id_student",
            "gender",
            "age_band",
            "highest_education",
            "imd_band",
            "studied_credits",
            "num_of_prev_attempts",
            "disability",
            "region"
        ]
    ]
    .drop_duplicates(subset="id_student")
    .copy()
)

print(f"Shape: {student_master.shape}")

student_master.head()
```

```python
# ==========================================================
# Validation
# ==========================================================

print("Unique students:", student_master["id_student"].nunique())
print("Total rows:", len(student_master))

assert student_master["id_student"].is_unique

student_master.info()
```

```python
# ==========================================================
# STEP 2A: Check Assessment Coverage
# ==========================================================

assessment_students = set(studentAssessment_clean["id_student"])

print("Students in student_master:",
      student_master["id_student"].nunique())

print("Students with assessments:",
      len(assessment_students))

print("Students without assessments:",
      len(set(student_master["id_student"]) - assessment_students))
```

```python
# ==========================================================
# STEP 2B: Assessment Feature Engineering
# ==========================================================

student_assessment_features = (
    studentAssessment_clean
    .groupby("id_student")
    .agg(
        avg_score=("score", "mean"),
        median_score=("score", "median"),
        max_score=("score", "max"),
        min_score=("score", "min"),
        score_std=("score", "std"),
        assessment_count=("score", "count"),
        avg_submission_day=("date_submitted", "mean"),
        first_submission=("date_submitted", "min"),
        last_submission=("date_submitted", "max")
    )
    .reset_index()
)

student_assessment_features.head()
```

```python
print(student_assessment_features.shape)

student_assessment_features.describe()
```

```python
student_assessment_features.isnull().sum()
```

```python
# Replace NaN standard deviations
student_assessment_features["score_std"] = (
    student_assessment_features["score_std"]
    .fillna(0)
)

student_master = student_master.merge(
    student_assessment_features,
    on="id_student",
    how="left"
)
```

```python
assessment_cols = [
    "avg_score",
    "median_score",
    "max_score",
    "min_score",
    "score_std",
    "assessment_count",
    "avg_submission_day",
    "first_submission",
    "last_submission"
]
student_master[assessment_cols] = (
    student_master[assessment_cols]
    .fillna(0)
)
```

```python
student_master.isnull().sum()
```

```python
student_master.head()
```

```python
# ==========================================================
# STEP 4A: Student Engagement Features
# ==========================================================

student_engagement_features = (
    studentVle_clean
    .groupby("id_student")
    .agg(
        total_clicks=("sum_click", "sum"),
        interaction_count=("sum_click", "count"),
        unique_resources=("id_site", "nunique"),
        active_days=("date", "nunique"),
        first_day=("date", "min"),
        last_day=("date", "max")
    )
    .reset_index()
)

# Derived features
student_engagement_features["active_span"] = (
    student_engagement_features["last_day"]
    - student_engagement_features["first_day"]
)

student_engagement_features["avg_clicks_per_interaction"] = (
    student_engagement_features["total_clicks"]
    / student_engagement_features["interaction_count"]
)

student_engagement_features["avg_clicks_per_day"] = (
    student_engagement_features["total_clicks"]
    / student_engagement_features["active_days"]
)
student_engagement_features["resource_revisit_ratio"] = (
    student_engagement_features["interaction_count"]
    / student_engagement_features["unique_resources"]
)

student_engagement_features.head()
```

```python
print(student_engagement_features.shape)

student_engagement_features.describe()
```

```python
engagement_cols = [
    "total_clicks",
    "interaction_count",
    "unique_resources",
    "active_days",
    "first_day",
    "last_day",
    "active_span",
    "avg_clicks_per_interaction",
    "avg_clicks_per_day",
    "resource_revisit_ratio"
]

student_master = student_master.merge(
    student_engagement_features,
    on="id_student",
    how="left"
)
```

```python
student_master[engagement_cols] = (
    student_master[engagement_cols]
    .fillna(0)
)
```

```python
student_master.isnull().sum()
```

```python
student_master.head()
```

```python
# ==========================================================
# STEP 5A: Aggregate Resource-Level Features
# ==========================================================

resource_engagement = (
    studentVle_clean
    .groupby("id_site")
    .agg(
        total_clicks=("sum_click", "sum"),
        interaction_count=("sum_click", "count"),
        unique_students=("id_student", "nunique")
    )
    .reset_index()
)

resource_engagement["avg_clicks_per_student"] = (
    resource_engagement["total_clicks"]
    / resource_engagement["unique_students"]
)

resource_engagement["avg_clicks_per_interaction"] = (
    resource_engagement["total_clicks"]
    / resource_engagement["interaction_count"]
)

resource_engagement["resource_revisit_rate"] = (
    resource_engagement["interaction_count"]
    / resource_engagement["unique_students"]
)

resource_engagement.head()
```

```python
# ==========================================================
# STEP 5B: Build Resource Master
# ==========================================================

resource_master = vle_clean.merge(
    resource_engagement,
    on="id_site",
    how="left"
)
```

```python
engagement_cols = [
    "total_clicks",
    "interaction_count",
    "unique_students",
    "avg_clicks_per_student",
    "avg_clicks_per_interaction",
    "resource_revisit_rate"
]

resource_master[engagement_cols] = (
    resource_master[engagement_cols]
    .fillna(0)
)

resource_master.head()
```

```python
print(resource_master.shape)

resource_master.isnull().sum()
```

```python
#Student_resource table
student_resource_interactions = (
    studentVle_clean
    .groupby(
        [
            "id_student",
            "id_site",
            "code_module",
            "code_presentation"
        ],
        as_index=False
    )
    .agg(
        total_clicks=("sum_click", "sum"),
        interaction_count=("sum_click", "count"),
        active_days=("date", "nunique"),
        first_interaction=("date", "min"),
        last_interaction=("date", "max")
    )
)

student_resource_interactions["interaction_span"] = (
    student_resource_interactions["last_interaction"]
    - student_resource_interactions["first_interaction"]
)

student_resource_interactions["avg_clicks_per_interaction"] = (
    student_resource_interactions["total_clicks"]
    / student_resource_interactions["interaction_count"]
)

student_resource_interactions["target"] = 1
```

```python
import random
random.seed(42)
```

```python
negative_rows = []

for (student, module, presentation), clicked_resources in student_clicked.items():

    all_resources = module_resources[(module, presentation)]

    candidate_negatives = list(
        all_resources - clicked_resources
    )

    if len(candidate_negatives) == 0:
        continue

    sampled_negatives = random.sample(
        candidate_negatives,
        min(
            len(clicked_resources),
            len(candidate_negatives)
        )
    )

    for resource in sampled_negatives:

        negative_rows.append({

            "id_student": student,
            "id_site": resource,
            "code_module": module,
            "code_presentation": presentation,
            "target": 0

        })
```

```python
negative_samples = pd.DataFrame(negative_rows)

negative_samples.head()
```

```python
import numpy as np
```

```python
def ndcg_at_k(group, k=10):

    group = group.sort_values(
        "prediction_score",
        ascending=False
    ).head(k)

    relevance = group["target"].values

    dcg = np.sum(
        relevance / np.log2(np.arange(2, len(relevance) + 2))
    )

    ideal = np.sort(relevance)[::-1]

    idcg = np.sum(
        ideal / np.log2(np.arange(2, len(ideal) + 2))
    )

    if idcg == 0:
        return 0

    return dcg / idcg
```

```python
def recommend(student_id, top_k=10):
    """
    Generate Top-K resource recommendations for a student.

    Parameters
    ----------
    student_id : int
        Student ID

    top_k : int
        Number of recommendations to return

    Returns
    -------
    pandas.DataFrame
        Top-K recommended resources with prediction scores
    """

    # --------------------------------------------------
    # 1. Find student's module & presentation
    # --------------------------------------------------

    student_course = (
        student_resource_interactions[
            student_resource_interactions["id_student"] == student_id
        ][
            ["code_module", "code_presentation"]
        ]
        .drop_duplicates()
    )

    if student_course.empty:
        raise ValueError(f"Student {student_id} not found.")

    code_module = student_course.iloc[0]["code_module"]
    code_presentation = student_course.iloc[0]["code_presentation"]

    # --------------------------------------------------
    # 2. Retrieve all resources in the course
    # --------------------------------------------------

    candidate_resources = resource_master[
        (resource_master["code_module"] == code_module)
        &
        (resource_master["code_presentation"] == code_presentation)
    ].copy()

    # --------------------------------------------------
    # 3. Remove resources already interacted with
    # --------------------------------------------------

    clicked_resources = set(
        student_resource_interactions.loc[
            student_resource_interactions["id_student"] == student_id,
            "id_site"
        ]
    )

    candidate_resources = candidate_resources[
        ~candidate_resources["id_site"].isin(clicked_resources)
    ].copy()

    if candidate_resources.empty:
        print("No unseen resources available.")
        return pd.DataFrame()

    # --------------------------------------------------
    # 4. Student features
    # --------------------------------------------------

    student_features = student_master[
        student_master["id_student"] == student_id
    ].copy()

    # --------------------------------------------------
    # 5. Preference features
    # --------------------------------------------------

    student_pref = student_preferences[
        student_preferences["id_student"] == student_id
    ].copy()

    # --------------------------------------------------
    # 6. Merge student + preference features
    # --------------------------------------------------

    student_info = student_features.merge(
        student_pref,
        on="id_student",
        how="left"
    )

    # --------------------------------------------------
    # 7. Repeat student row
    # --------------------------------------------------

    student_info = pd.concat(
        [student_info] * len(candidate_resources),
        ignore_index=True
    )

    # --------------------------------------------------
    # 8. Combine with resource features
    # --------------------------------------------------

    candidate_data = pd.concat(
        [
            student_info.reset_index(drop=True),
            candidate_resources.reset_index(drop=True)
        ],
        axis=1
    )

    # --------------------------------------------------
    # 9. Rename columns
    # --------------------------------------------------

    candidate_data.columns = [
        "id_student",
        "gender",
        "age_band",
        "highest_education",
        "imd_band",
        "studied_credits",
        "num_of_prev_attempts",
        "disability",
        "region",
        "avg_score",
        "median_score",
        "max_score",
        "min_score",
        "score_std",
        "assessment_count",
        "avg_submission_day",
        "first_submission",
        "last_submission",
        "student_total_clicks",
        "student_interaction_count",
        "student_unique_resources",
        "student_active_days",
        "student_first_day",
        "student_last_day",
        "student_active_span",
        "student_avg_clicks_per_interaction",
        "student_avg_clicks_per_day",
        "student_resource_revisit_ratio",
        "pref_forumng",
        "pref_other",
        "pref_oucollaborate",
        "pref_oucontent",
        "pref_page",
        "pref_quiz",
        "pref_resource",
        "pref_subpage",
        "pref_url",
        "id_site",
        "code_module",
        "code_presentation",
        "activity_type",
        "resource_total_clicks",
        "resource_interaction_count",
        "resource_unique_students",
        "resource_avg_clicks_per_student",
        "resource_avg_clicks_per_interaction",
        "resource_revisit_rate"
    ]

    # --------------------------------------------------
    # 10. Save recommendation info
    # --------------------------------------------------

    recommendation_info = candidate_data[
        [
            "id_student",
            "id_site",
            "activity_type"
        ]
    ].copy()

    # --------------------------------------------------
    # 11. Build model input
    # --------------------------------------------------

    candidate_X = candidate_data[
        feature_columns
    ].copy()

    # --------------------------------------------------
    # 12. Encode categorical features
    # --------------------------------------------------

    candidate_X[categorical_features] = encoder.transform(
        candidate_X[categorical_features]
    )

    # --------------------------------------------------
    # 13. Predict probabilities
    # --------------------------------------------------

    recommendation_info["prediction_score"] = (
        lgb_model.predict_proba(candidate_X)[:, 1]
    )

    # --------------------------------------------------
    # 14. Sort recommendations
    # --------------------------------------------------

    recommendations = (
        recommendation_info
        .sort_values(
            by="prediction_score",
            ascending=False
        )
        .head(top_k)
        .reset_index(drop=True)
    )

    recommendations.insert(
        0,
        "Rank",
        range(1, len(recommendations) + 1)
    )

    return recommendations
```
